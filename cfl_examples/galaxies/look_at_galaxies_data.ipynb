From Eric Huff: This is a set of examples of galaxy images and maps of related physical quantities, as I’d shown on our last call. There are 413 examples, packaged into two arrays called “inputs” and “targets". Each row is a pair of 51x51 pixel 2D images, each of which has been flattened into a single 2601-element array prior to concatenation. To get an array with one of the images, you can do the following:

```python
import numpy as np
import matplotlib.pyplot as plt

# load data
npzfile = np.load('illustris_data_with_selection.npz')                                                                                                     
print('Data shape: ', npzfile['targets'].shape)                                                                                                            
```

```python
# pull out the 8th observation
iim1 = npzfile['inputs'][8,:2601].reshape((51,51))
iim2 = npzfile['inputs'][8,2601:].reshape((51,51))                                                               
tim1 = npzfile['targets'][8,:2601].reshape((51,51))
tim2 = npzfile['targets'][8,2601:].reshape((51,51))
```

```python
# plot
fig,ax = plt.subplots(2,2, figsize=(6,7))
ax[0,0].imshow(iim1)
ax[0,0].set_title('Input 1')
ax[0,1].imshow(iim2)
ax[0,1].set_title('Input 2')
ax[1,0].imshow(tim1)
ax[1,0].set_title('Target 1')
ax[1,1].imshow(tim2)
ax[1,1].set_title('Target 2')
plt.show()
```

```python
print('Log Scale')
fig,ax = plt.subplots(1,2)
ax[0].imshow(np.log(tim1))
ax[0].set_title('Target 1')
ax[1].imshow(np.log(tim2))
ax[1].set_title('Target 2')
plt.show()
```

```python
def print_range(label, data):
    print("{}: {}, {}".format(label, round(np.min(data),3), round(np.max(data),3)))
```

```python
print_range('input 0', iim1)
print_range('input 1', iim2)
print_range('target 0', tim1)
print_range('target 1', tim2)
```
